```scala
import org.apache.spark.sql.functions.{explode, split, col}
```

```scala
val spark = SparkSession.builder().appName("User Rating History").master("local[*]").getOrCreate()
val sparkConext = spark.sparkContext
```

```scala
val rating_gcs_path = "gs://artifacts_spark_jobs/rating.csv"


val rating_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(rating_gcs_path)
```

```scala
val validRatingsDF = rating_df.filter(
  col("userId").isNotNull &&
  col("movieId").isNotNull &&
  col("rating").isNotNull
)

val limitedDF = validRatingsDF.limit(1000) // limiting the records as the kernel is able to process large amount of data
```

```scala
val outputBasePath = "hdfs:/user/hdfs/CaseStudies/userData/"
limitedDF.write
  .partitionBy("userId")  // Partition the data by userId
  .format("parquet")      // Use Parquet format
  .mode("overwrite")      // Overwrite if exists
  .save(outputBasePath)
```
