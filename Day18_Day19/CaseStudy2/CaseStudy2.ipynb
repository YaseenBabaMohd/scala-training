```scala
import org.apache.spark.sql.{SparkSession, DataFrame}
import org.apache.spark.sql.functions._
```

```scala
val spark = SparkSession.builder()
             .appName("Caching Example")
             .master("local[*]")
             .getOrCreate()

val sparkContext = spark.sparkContext
```

```scala
val transactionsLocation = "gs://artifacts_spark_jobs/movie_metadata/yaseen/transaction_logs.csv"
val usersData = "gs://artifacts_spark_jobs/movie_metadata/yaseen/user_details.csv"
```

```scala
val transactionsData = spark.read.format("csv").option("header","true").option("inferSchema","true").load(transactionsLocation)
```

```scala
val userData = spark.read.format("csv").option("header", "true").option("inferSchema","true").load(usersData)
```

```scala
 val startWithoutCache = System.currentTimeMillis()
    for (_ <- 1 to 5) {
      transactionsData.groupBy("transaction_type").agg(sum("amount").as("total_sales")).show()
    }
val endWithoutCache = System.currentTimeMillis()
    println(s"Time without caching: ${endWithoutCache - startWithoutCache} ms")
```

```scala
transactionsData.cache() // Alternatively: salesData.persist(StorageLevel.MEMORY_AND_DISK)
 val startWithoutCache = System.currentTimeMillis()
    for (_ <- 1 to 5) {
      transactionsData.groupBy("transaction_type").agg(sum("amount").as("total_sales")).show()
    }
val endWithoutCache = System.currentTimeMillis()
    println(s"Time without caching: ${endWithoutCache - startWithoutCache} ms")
```
